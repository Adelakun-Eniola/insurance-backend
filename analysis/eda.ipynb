```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style ='whitegrid')
```

```python
df = pd.read_csv('../data/processed/cleaned_dataset_.csv')
```

```python
print("\n: Dataset Overview:")
print(df.info())
print("\n: Null values")
print(df.isnull().sum())
```

```python
print("\n: Summary Overview:")
print(df.describe())
```

```python
cat_columns = ['sex', 'smoker']
for col in cat_columns:
    print(f"\n[INFO] Value counts for {col}:")
    print(df[col].value_counts())
    # print(df.columns.tolist())
    sns.countplot(x=col, data=df)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python

sns.histplot(df['medical charges'], kde = True)
plt.title("Distribution of Insurance Charges")
plt.xlabel("Charges")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.scatterplot(x='age', y='medical charges', data=df)
plt.title("Charges vs Age")
plt.show()
```

```python
sns.scatterplot(x='bmi', y='medical charges', data=df, hue='smoker')
plt.title("Charges vs BMI (colored by Smoker)")
plt.show()
```

```python
sns.boxplot(x='smoker', y='medical charges', data=df)
plt.title("Charges by Smoking Status")
plt.show()
```

```python
sns.boxplot(x='region_southwest', y='medical charges', data=df)
plt.title("Charges by Region")
plt.show()
```
